```python
import numpy as np
import cv2

import matplotlib.pyplot as plt

# %matplotlib inline
plt.rcParams['figure.figsize'] = [100,80]

# %matplotlib notebook
```

```python
img = cv2.imread('sample.png')

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def birdview(img,shrink_ratio):
    SKYLINE = 85
    height,width = img.shape[:2]
    
    roi = img.copy()
#     mask = np.zeros((height,width),dtype=np.uint8)
    
    cv2.rectangle(roi,(0,0),(width,SKYLINE),0,-1)
#     roi = cv2.bitwise_and(img,img,mask=mask)
    
    src_pts = np.float32([[0,SKYLINE],[width,SKYLINE],[0,height],[width,height]])
    
    
    dst_width = 480
    dst_height = 480
    dsize = (dst_width,dst_height)

    dst_pts = np.float32([[0,0],[dst_width,0],[dst_width*shrink_ratio,dst_height],[dst_width*(1-shrink_ratio),dst_height]])

    M = cv2.getPerspectiveTransform(src_pts,dst_pts)

    dst = cv2.warpPerspective(roi,M,dsize)

    return dst
```

Use HLS (hue, lightness, saturation) color space for more accurate gradient detection.

<img src='hls_color_model.png'>

Find lightness channel gradient:

```python
roi = img.copy()
height,width=roi.shape[:2]
cv2.rectangle(roi,(0,0),(width,85),0,-1)

hls = cv2.cvtColor(roi,cv2.COLOR_RGB2HLS)
h_channel = hls[:,:,0]
l_channel = hls[:,:,1]
s_channel = hls[:,:,2]

# Sobel both x and y directions
sobel = cv2.Sobel(l_channel,cv2.CV_64F,1,1)
# sobel = cv2.Laplacian(l_channel,cv2.CV_64F)
abs_sobel = np.absolute(sobel) # absolute all negative gradient values

# cv2.imwrite('gradient.png',abs_sobel)

plt.imshow(abs_sobel,cmap='gray')
plt.show()
```

Threshold lightness gradient for significant changes

```python
# %matplotlib notebook
ax = plt.subplot(221)
ax.set_title('lightness')
plt.imshow(l_channel,cmap='gray')

l_ret,l_thresh = cv2.threshold(abs_sobel,75,255,cv2.THRESH_BINARY)
ax = plt.subplot(222)
ax.set_title('l_thresh')
plt.imshow(l_thresh,cmap='binary')

ax = plt.subplot(223)
ax.set_title('saturation')
plt.imshow(s_channel,cmap='gray')

# threshold saturation channel
s_sobel = cv2.Sobel(s_channel,cv2.CV_64F,1,1)
abs_s_sobel = np.absolute(s_sobel)
s_ret,s_thresh = cv2.threshold(abs_s_sobel,100,255,cv2.THRESH_BINARY)

ax = plt.subplot(224)
ax.set_title('s_thresh')
plt.imshow(s_thresh,cmap='binary')
# ax.set_title('abs_s_sobel')
# plt.imshow(abs_s_sobel,cmap='gray')

plt.show()
```

Get bird's eye view with multiple shrink ratios

Shrink ratio should be between 40 and 47 percent

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [32,24]

plot_rows=3
plot_cols=3
for i in range(8):
    
    ratio = (40+i)/100.0
    
    bird_view = birdview(l_thresh,ratio)
    
    label = '{}%'.format(int(ratio*100))
    
    plt.subplot(plot_rows, plot_cols, i+1)
    plt.imshow(bird_view,cmap='binary')
    
plt.show()
```
